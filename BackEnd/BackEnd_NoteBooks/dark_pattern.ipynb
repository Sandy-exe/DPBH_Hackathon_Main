```python
from fastapi import FastAPI
from pydantic import BaseModel
import pickle
import json
import uvicorn
from pyngrok import ngrok
from fastapi.middleware.cors import CORSMiddleware
import nest_asyncio
```

```python
app = FastAPI()
```

```python
origins = ["*"]

app.add_middleware(
    CORSMiddleware,
    allow_origins=origins,
    allow_credentials=True,
    allow_methods=["*"],
    allow_headers=["*"],
)
```

```python
class model_input(BaseModel):
  Query : str
```

```python
@app.post('/llama')
def dark_predd(input_parameters : model_input):

    input_data = input_parameters.json()
    input_dictionary = json.loads(input_data)

    query = input_dictionary['Query']

    value = pipe(query)
    if value[0]['label'] == "Dark_Pattern":
      return str({"Dark_pattern":value[0]['score']})
    elif value[0]['label'] == "Not_Dark_Pattern":
      return str({"Not_Dark_pattern":value[0]['score']})
```
