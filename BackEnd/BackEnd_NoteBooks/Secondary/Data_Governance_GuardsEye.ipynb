```python
service_context = ServiceContext.from_defaults(
    chunk_size=1024,
    llm=llm,
    embed_model=embed_model
)
```

```python
index = VectorStoreIndex.from_documents(documents, service_context=service_context)
```

```python
from llama_index.core import ChatPromptTemplate
from llama_index.core.llms import ChatMessage, MessageRole
```

```python
TEXT_QA_SYSTEM_PROMPT = ChatMessage(
    content=(
        "You are an AI assistant trained to Summarize the flipkart data governance laws. Go through the flipkart policy file and gdpr to answer the question."
        "Go through the data throughly before answering"
        "Always answer the query using the provided context information but say it out"
        "and not prior knowledge.\n"
        "Some rules to follow:\n"
        "1. directly reference the given context in your answer.\n"
        "2. Avoid statements like 'Based on the context, ...' or "
        "'The context information ...' or anything along "
        "those lines."
    ),
    role=MessageRole.SYSTEM,
)

TEXT_QA_PROMPT_TMPL_MSGS = [
    TEXT_QA_SYSTEM_PROMPT,
    ChatMessage(
        content=(
            "Context information is below.\n"
            "---------------------\n"
            "{context_str}\n"
            "---------------------\n"
            "Given the context information and not prior knowledge, "
            "Answer the query.\n"
            "Query: {query_str}\n"
            "Answer: "
        ),
        role=MessageRole.USER,
    ),
]

CHAT_TEXT_QA_PROMPT = ChatPromptTemplate(message_templates=TEXT_QA_PROMPT_TMPL_MSGS)
```

```python
print(CHAT_TEXT_QA_PROMPT)
```

```python
from contextvars import Context
query_engine = index.as_query_engine(streaming = True, text_qa_template=CHAT_TEXT_QA_PROMPT )
```

```python
query = "Answer in 1 line, is flipkart adhereing to the data governance laws? And then summarize the entire data governance policy in 5 lines with max 10 words in each line"
# default prompt
response = query_engine.query(query)
response.print_response_stream()



```
